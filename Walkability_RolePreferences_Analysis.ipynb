```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
```

```python
# name of the files in the folder
for dirname, _, filenames in os.walk('/content/drive/MyDrive/Walkability'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# data fetched
data = pd.read_csv("/content/drive/MyDrive/Walkability/PrimaryTaskAllocation.csv")


# checking the train_data columns
data.head()
```

```python
# deleted redundant column
del data["Timestamp"]
del data["Email Address"]
```

```python
data.head()
```

```python
data.rename(
    columns={"Which sector(s) do you prefer primarily?\n\nNote that data collection (walking) is everyone's duty by default.": "Role Preferences"},
    inplace=True,
)

data.head()
```

```python
gk = data.groupby('Role Preferences')
gk.first()
```

```python
# Iterate over columns using DataFrame.iteritems()
for (colname,colval) in data.iteritems():
    if colname == 'Role Preferences':
      print(colval.values)
```

```python
data.head()
```

```python
data['Data Cleaning'] = 0
data['Data Modelling'] = 0
data['Documentation'] = 0
data['UI/UX'] = 0
data['Web App (Frontend)'] = 0
data['API (Backend)'] = 0

data.head()
```

```python
data.info()
```

```python
for item, frame in data['Role Preferences'].iteritems():
    if pd.notnull(frame):
        choices = frame.split(", ")
        for choice in choices:
          data.loc[item, choice] = 1
```

```python
data = data.sort_values('Name')
data.head()
```

```python
# save DataFrame to a CSV file
data.to_csv('RolePreferenceDistribution.csv', index=False)
print("Your submission was successfully saved!")
```
